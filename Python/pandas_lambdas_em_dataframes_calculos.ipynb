```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
      'a':[1,np.nan,7,3,np.nan],
      'b':[np.nan,5,np.nan,2,6], 
      })

df
```

```python
df['a_'] = df.apply(lambda row: row['b'] if np.isnan(row['a']) else row['a'],axis=1)
df['b_'] = df.apply(lambda row: row['a'] if np.isnan(row['b']) else row['b'],axis=1)
df
```
