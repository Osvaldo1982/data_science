```python
import pandas as pd
```

# Importando o dataset de exemplo

```python
df = pd.read_csv('03_tables_groupby_melt_exemplos.csv',encoding = 'ISO-8859-1', delimiter=';',low_memory = False)
```

```python
df.info()
```

## Criando um dataframe a partir de um groupby

```python
df['contador'] = 1

list_types = ['contador','age_years','bmi_kg_m2','height_cm','weight_kg','hdl_mg_dl','ldl_mg_dl','non_hdl_mg_dl',
              'total_cholesterol_mg_dl','trygliceride_mg_dl','fast_blood_sugar_mg_dl','systolic_bp_mm_hg',
              'diastolic_bp_mm_hg']

cluster_means = pd.DataFrame( df.groupby(['cluster']).agg({
'contador':'sum',
'age_years':'mean',
'bmi_kg_m2':'mean',
'height_cm':'mean',
'weight_kg':'mean',
'hdl_mg_dl':'mean',
'ldl_mg_dl':'mean',
'non_hdl_mg_dl':'mean',
'total_cholesterol_mg_dl':'mean',
'trygliceride_mg_dl':'mean',
'fast_blood_sugar_mg_dl':'mean',
'systolic_bp_mm_hg':'mean',
'diastolic_bp_mm_hg':'mean',
})).reset_index()
cluster_means
```

# Uso do melt para criar tabelas verticais

## Fazendo você mesmo

```python
df_aux = pd.DataFrame(df.groupby(['cluster']).agg({'bmi_kg_m2':'mean'}))
df_aux = df_aux.rename(columns = {'bmi_kg_m2':'type_values'})
df_aux['type'] = 'bmi_kg_m2'
df_aux
```

## Usando o melt

```python
cluster_means_vertical = pd.melt(cluster_means, id_vars=['cluster'], value_vars = list_types)
cluster_means_vertical
```
