```python
import pandas as pd
import numpy as np
```

```python
pd.set_option('display.max_columns', None)
```

```python
data_train = pd.read_csv('train.csv', index_col='NU_INSCRICAO')
data_train.head()
```

```python
data_test = pd.read_csv('test.csv', index_col='NU_INSCRICAO')
data_test.head()
```

```python
data_test.columns
```

```python
train_cols = ['CO_UF_RESIDENCIA', 'NU_IDADE',
       'TP_SEXO', 'TP_COR_RACA', 'TP_NACIONALIDADE', 'TP_ST_CONCLUSAO',
       'TP_ANO_CONCLUIU', 'TP_ESCOLA', 'TP_ENSINO', 'IN_TREINEIRO',
       'TP_DEPENDENCIA_ADM_ESC', 'IN_BAIXA_VISAO', 'IN_CEGUEIRA', 'IN_SURDEZ',
       'IN_DISLEXIA', 'IN_DISCALCULIA', 'IN_SABATISTA', 'IN_GESTANTE',
       'IN_IDOSO', 'TP_PRESENCA_CN', 'TP_PRESENCA_CH', 'TP_PRESENCA_LC',
       'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'TP_LINGUA',
       'TP_STATUS_REDACAO', 'NU_NOTA_COMP1', 'NU_NOTA_COMP2', 'NU_NOTA_COMP3',
       'NU_NOTA_COMP4', 'NU_NOTA_COMP5', 'NU_NOTA_REDACAO']
```

```python
# Drop rows where the target is NaN
data_train.dropna(subset = ['NU_NOTA_MT'], inplace = True)
# Replace NaN values with -1 otherwise
data_train.fillna(-1, inplace=True)
```

```python
X = data_train[train_cols]
y = data_train["NU_NOTA_MT"]

# One-hot encode the data
X = pd.get_dummies(X, drop_first=True)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
X = sc_X.fit_transform(X)
```

```python
# Applying k-Fold Cross Validation for different methods

from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor()

from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression(fit_intercept=False)

#from xgboost import XGBoostRegressor
#xgb_reg = XGBoostRegressor()


models = ['rf_reg', 'lin_reg']
```

```python
from sklearn.model_selection import cross_val_score
for model in models:
    regressor = eval(model)
    scores = cross_val_score(estimator=regressor, X=X, y=y, cv = 5)
    print(f'{model}: scores mean = {scores.mean():.2f} +- {scores.std():.2f}')
```

```python
X_test = data_test[train_cols].fillna(-1)
X_test = pd.get_dummies(X_test, drop_first=True)
X_test.head()
```

```python
X_test = sc_X.transform(X_test)
y_pred = rf_reg.fit(X, y).predict(X_test)
answer = pd.DataFrame(data={'NU_NOTA_MT': y_pred}, index=data_test.index)
answer[answer['NU_NOTA_MT'] < 0] = np.nan
answer.head()
```
