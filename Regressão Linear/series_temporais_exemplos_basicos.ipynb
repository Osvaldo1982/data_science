```python
from __future__ import print_function
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
med = 15.5
dp = 8.2
sDays = np.arange('2001-01', '2016-12', dtype='datetime64[D]')
nDays = len(sDays)

s1 = np.random.gumbel(loc=med,scale=dp,size=nDays)
s1[s1 < 0] = 0

dfSint = pd.DataFrame({'Q':s1},index=sDays)
dfSint.plot()
```

```python
nsample = 1024

## simulate a simple sinusoidal function
x1 = np.linspace(0, 100, nsample)

sigma = 0.3

y =np.sin(x1) + sigma * np.random.normal(size=nsample)
fig, ax = plt.subplots()

ax.plot(x1, y, label="Data")
```
